```python
import datetime
from bs4 import BeautifulSoup
import requests
import string
import pandas as pd
from collections import Counter
```

```python
re = requests.get("https://kun.uz/uz/news/list")
frontend = re.text
soup = BeautifulSoup(frontend, 'html.parser')

source_urls = []
for row in soup.find_all('a', class_="daily-block l-item"):
    source_urls.append("https://kun.uz" + row['href'])
```

```python
source_urls
```

```python
def get_article(url):
    re = requests.get(url)
    frontend = re.text
    soup = BeautifulSoup(frontend, 'html.parser')

    content = str()
    for p in soup.find_all('p'):
        content += p.text
    date_time = datetime.datetime.now().strftime("%d/%m/%Y, %H:%M:%S")

    return date_time, content
```

```python
access_datetimes = []
contents = []

for url in source_urls:
    date_time, content = get_article(url)
    access_datetimes.append(date_time)
    contents.append(content.lower())
```

```python
data_dict = {'source_url': source_urls, 'access_datetime': access_datetimes, 'content': contents}
```

```python
data = pd.DataFrame(data_dict)
```

```python
data
```

```python
words = [content.translate(content.maketrans('', '', string.punctuation+"\t\n\r\x0b\x0c\xa0")).split(' ') for content in contents]
```

```python
words
```

```python
data['word'] = words
```

```python
data
```

```python
words_number = [Counter(word) for word in words]
```

```python
data['words_number'] = words_number
```

```python
data
```

```python
data.to_csv('kun_uz.csv', sep='|', index=False)
```
